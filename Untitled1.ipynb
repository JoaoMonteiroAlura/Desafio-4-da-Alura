```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Trilhas2B-Desafio 4  - Churn.csv')
```

```python
print("Formato do dataset:", df.shape)
```

```python
df.columns = df.columns.str.lower()
```

```python
print(df.isnull().sum())
```

```python
df['salario anual'] = df['salario anual'].fillna(df['salario anual'].mean())
```

```python
df['genero'] = df['genero'].fillna(df['genero'].mode()[0])
```

```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
def tratar_outliers_iqr(df, colunas):
    for coluna in colunas:
        if pd.api.types.is_numeric_dtype(df[coluna]):
            Q1 = df[coluna].quantile(0.25)
            Q3 = df[coluna].quantile(0.75)
            IQR = Q3 - Q1
            limite_inferior = Q1 - 1.5 * IQR
            limite_superior = Q3 + 1.5 * IQR
            df.loc[:, coluna] = df[coluna].apply(lambda x:
                limite_inferior if x < limite_inferior
                else limite_superior if x > limite_superior
                else x
            )
    return df

colunas_numericas = ['salario anual', 'idade', 'saldo na conta', 'bens', 'pontos']
df = tratar_outliers_iqr(df, colunas_numericas)
```

```python
df = df.copy()
df['genero'] = df['genero'].str.strip().str.lower()
df['genero'] = df['genero'].replace({
    'mas': 'masculino',
    'masculino': 'masculino',
    'fem': 'feminino',
    'feminino': 'feminino'
})
```

```python
df.columns = df.columns.str.lower().str.strip()
```

```python
print(df.shape)
```

```python
contagem_genero_estado = df.groupby(['genero', 'estado']).size().reset_index(name='contagem')
print(contagem_genero_estado)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=contagem_genero_estado, x='estado', y='contagem', hue='genero')
plt.title('Contagem de registros por Estado e Gênero')
plt.show()
```

```python
churn_genero = df.groupby('genero')['saiu'].mean().reset_index()
print(churn_genero)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=churn_genero, x='genero', y='saiu')
plt.ylabel('Taxa de Saída (Churn)')
plt.title('Taxa de Saída por Gênero')
plt.show()
```

```python
acima_40 = df[df['idade'] > 40]['saldo na conta']
media_acima_40 = acima_40.mean()
mediana_acima_40 = acima_40.median()

print(f"Média do saldo (idade > 40): {media_acima_40:.2f}")
print(f"Mediana do saldo (idade > 40): {mediana_acima_40:.2f}")
```
